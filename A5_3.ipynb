```python
import numpy as np
import matplotlib.pyplot as plt
import sys
```

```python
infile = open('housing.csv')
rawdata = infile.readlines()
print(rawdata[0])
infile.close
print("Size of data is {} bytes".format(sys.getsizeof(rawdata)))
```

```python
print("Data imported as {} of {}s".format(type(rawdata)),type(rawdata[0]))
```

```python
data = []
for line in rawdata:
    data.append(line.strip().split(','))
print("Dimensions of housing data is {} x {}".format(len(data[0]),len(data)))
```

```python
popIndex = data[0].index('population')
pop = []
for line in data[1:]:
    pop.append(int(float(line[popIndex])))
print(pop[:10])
plt.hist(pop, bins=20)
plt.xlabel('Population')
plt.show()
```

```python
p = {}
for line in data[1:]:
    if line[-1] in p:
        p[line[-1]] += 1
    else:
        p[line[-1]] = 1

plt.bar(range(len(p)), list(p.values()), align='center')
plt.xticks(range(len(p)), list(p.keys()))
plt.xlabel('Proximity to Ocean')
plt.show()
```
